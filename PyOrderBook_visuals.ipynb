### Directional Signal Explanation

Below, we examine the stochastic nature of the data, what you can see is:

1.   The returns, hard-labelled as -1, 0, or 1, have many fluctuations
2.   Specifically with this dataset, we can see that there are about 36 "shifts" in the first graph.
3.   On the other hand, the second graph that uses our directional signal labels only has 24 "shifts". Furthermore, the shifts here are less impactful as they might go from -1 to 0 and back to -1 more often, instead of going from -1 to 0, then to 1



```python
stoch = [0]

for i in range(1, len(data)):
  if data['mid_price'][i] > data['mid_price'][i-1]:
    stoch.append(1)
  elif data['mid_price'][i] < data['mid_price'][i-1]:
    stoch.append(-1)
  else:
    stoch.append(0)


plt.style.use("cyberpunk")
plt.plot(stoch)
```

```python
plt.plot(data['dir'])
```

By creating directional signals, that are held to an alpha value and set on a k horizon, we can reduce the stochastic nature of the data, which provides a feature to the models that is probably more helpful in predicting the future

Comparing these on a plot of the bitcoin price, using conditional formatting to 



*   We can see that using a unsmoothed version of direction leads to a very stochastic chart with many changes
*   On the contrary, using the smoothed directional signal gives us a better understanding of the pricemovements as a whole. Because this is time series data, we want to examine ***trends***!



indicate these directional signals:

```python
#Tracking both Bid and Ask Price
data_n = data.copy()
data_n['stoch'] = stoch


plt.style.use("Solarize_Light2")
plt.rcParams['figure.figsize'] = (16,8)

ax = data_n['mid_price'].plot()

ax.set_title('Bitcoin BitMex ', fontsize = 18,fontweight = 'bold', color = 'black')

ax.set_xlabel('Timestamp index', fontsize = 15, color = 'black',fontweight = 'bold')
ax.set_ylabel('Prices', fontsize = 15, color = 'black',fontweight = 'bold')

ax.pcolorfast(ax.get_xlim(), ax.get_ylim(),
              data_n['stoch'].values[np.newaxis],
              cmap='RdYlGn', alpha=0.3)

plt.yticks(fontsize = 12, color = 'black',fontweight = 'bold')
plt.xticks(fontsize = 12, rotation = 90, color = 'black',fontweight = 'bold')


plt.show()

```

### Liquidity/Market Depth Visualizations

```python
def clean_data(data_buys, data_sells):

    buys_df = pd.DataFrame(data_buys)
    buys_df = buys_df.rename(columns = {0: 'price', 1: 'size', 2: 'timestamp'})
    buys_df = buys_df.sort_values(by = ['price'], ascending = False, axis = 0, ignore_index = True)
    buys_df['Liquidity'] = buys_df['price'] * buys_df['size']
    buys_df['Liquidity_running'] = buys_df['Liquidity'].cumsum(axis = 0)

    # print(buys_df)

    sells_df = pd.DataFrame(data_sells)
    sells_df = sells_df.rename(columns = {0: 'price', 1: 'size', 2: 'timestamp'})
    sells_df['Liquidity'] = sells_df['price'] * sells_df['size']
    sells_df['Liquidity_running'] = sells_df['Liquidity'].cumsum(axis = 0)

    # print(sells_df)

    return buys_df, sells_df

buys = pd.read_csv("/content/12_26_2022,01_20_08buys.csv", header = None)
sells = pd.read_csv("/content/12_26_2022,01_20_08sells.csv", header = None)
```

```python
buys_df, sells_df = clean_data(buys, sells)
```

```python
plt.style.use("cyberpunk")

fig, ax = plt.subplots()
ax.set_title('XBT-USD' + ' ' + 'order book on' + ' '+ str('BitMex') +', '+ 'on 2022-12-26 at 1:20am')#timestamp variable here

# bid side
sns.ecdfplot(x="price", weights="size", stat="count", complementary=True,
        data=buys_df.loc[buys_df['Liquidity_running'] < sum(buys_df['Liquidity']) / 10], ax=ax, color='g')
# ask side
sns.ecdfplot(x="price", weights="size", stat="count",
        data=sells_df.loc[sells_df['Liquidity_running'] < sum(sells_df['Liquidity']) / 10], ax=ax, color='r')

ax.set_xlabel("Price")
ax.set_ylabel("Amount")
```

So It seems like the sell side has a lot of pressure. 10% of the sell side is huge in comparison to the buy side, which makes sense since there is probably a lot of sellers with orders that will never be filled.
